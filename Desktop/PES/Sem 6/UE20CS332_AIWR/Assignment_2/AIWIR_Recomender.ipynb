```python
import numpy as np
from rake_nltk import Rake
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df
```

```python
import string
for i in range(len(df['text'])):
    df['text'][i] = df['text'][i].translate(str.maketrans('', '', string.punctuation))

df
```

```python
df['Key_words'] = ''   # initializing a new column
r = Rake()   # using Rake to remove stop words
```

```python
df
```

```python
for index, row in df.iterrows():
    r.extract_keywords_from_text(row['text'])   # to extract key words 
    key_words_dict_scores = r.get_word_degrees()    # to get dictionary with key words and their similarity scores
    row['Key_words'] = list(key_words_dict_scores.keys())
```

```python
df
```

```python
df['artist'] = df['artist'].str.lower()
df
```

```python
df['Bag_of_words'] = ''
# # columns = ['artist','Key_words']

# # df["Bag_of_words"] = df['artist']+" "+ df["Key_words"].astype(str)
# # df["Bag_of_words"] = df[["artist", "Key_words"]].apply(" ".join, axis=1)
temp = " "
for i in range(len(df['text'])):
    temp = ' '.join([str(elem) for elem in df["Key_words"][i]])
    df['Bag_of_words'][i] = df['artist'][i] + " " + temp

```

```python
df
```

```python
# df['Bag_of_words'] = df['Bag_of_words'].str.strip().str.replace(' ', ' ').str.replace('  ', ' ')

df_new = df[['song','Bag_of_words']]
df_new
```

```python
df_1 = df_new.iloc[:int(len(df_new)/2),:]
df_2 = df_new.iloc[int(len(df_new)/2):,:]
```

```python
df_1
```

```python
count = CountVectorizer()
count_matrix = count.fit_transform(df_1['Bag_of_words'])
count_matrix
cosine_sim_1 = cosine_similarity(count_matrix, count_matrix)
print(cosine_sim_1)
```

```python
count = CountVectorizer()
count_matrix = count.fit_transform(df_2['Bag_of_words'])
count_matrix
cosine_sim_2 = cosine_similarity(count_matrix, count_matrix)
print(cosine_sim_2)
```

```python
cosine_sim = np.concatenate((cosine_sim_1,cosine_sim_2))
cosine_sim
```

```python
print(cosine_sim)
```

```python
indices = pd.Series(df['song'])
```

```python
def recommend(title, cosine_sim = cosine_sim):
    recommended_movies = []
    idx = indices[indices == title].index[0]   # to get the index of the movie title matching the input song
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)   # similarity scores in descending order
    top_5_indices = list(score_series.iloc[1:6].index)   # to get the indices of top 6 most similar songs
    # [1:6] to exclude 0 (index 0 is the input song itself)
    
    for i in top_5_indices:   # to append the titles of top 10 similar songs to the recommended_movies list
        recommended_movies.append(list(df_new['song'])[i])
        
    return recommended_movies
```

```python
recommend('As Good As New')
```
